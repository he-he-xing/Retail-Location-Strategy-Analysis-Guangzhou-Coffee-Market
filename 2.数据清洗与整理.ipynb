**计算目标品牌（星巴克）内部缓冲区重叠度：**

```python
buffers = geometry_target[['id', 'buffer']].copy()
buffers = buffers.set_geometry('buffer')
```

```python
#自身空间连接，找出所有相交对（包括自身）
intersect_pairs = gpd.sjoin(buffers, buffers, how='inner', predicate='intersects')
```

```python
# 过滤掉自身
intersect_pairs_screen=intersect_pairs[intersect_pairs['id_left']!=intersect_pairs['id_right']].copy()
```

```python
#计算每对交集的面积
def compute_intersection_area(row):
    buf_left = buffers.loc[buffers['id'] == row['id_left'], 'buffer'].values[0]
    buf_right = buffers.loc[buffers['id'] == row['id_right'], 'buffer'].values[0]
    return buf_left.intersection(buf_right).area

intersect_pairs_screen['area'] = intersect_pairs_screen.apply(compute_intersection_area, axis=1)
```

```python
# 按id_left分组求和（每个门店与其他门店的缓冲区重叠总面积）
overlap_sum = intersect_pairs_screen.groupby('id_left')['area'].sum().reset_index()
overlap_sum.columns = ['id', 'overlap_area']
```

```python
geometry_target = geometry_target.merge(overlap_sum, on='id', how='left').fillna(0)
```

```python
# 计算缓冲区面积
geometry_target['buffer_area'] = geometry_target['buffer'].area
```

```python
# 计算缓冲区重叠度
geometry_target['overlap_ratio'] = geometry_target['overlap_area'] / geometry_target['buffer_area']
```

**计算目标门店与竞品（瑞幸咖啡和库迪咖啡）的缓冲区重叠度：**

```python
geometry_competitor = geometry_competitor.reset_index(drop=False).rename(columns={'index': 'id'})
```

```python
buffers_competitor = geometry_competitor[['id', 'buffer']].copy()
buffers_competitor = buffers_competitor.set_geometry('buffer')
```

```python
intersect_pairs_comp = gpd.sjoin(buffers, buffers_competitor, how='left', predicate='intersects')
intersect_pairs_comp_filtered = intersect_pairs_comp.dropna(subset=['index_right']).copy()
```

```python
def compute_comp_intersection_area(row):
    buf_target = buffers.loc[buffers['id'] == row['id_left'], 'buffer'].values[0]
    buf_competitor = buffers_competitor.loc[buffers_competitor['id'] == row['id_right'], 'buffer'].values[0]
    return buf_target.intersection(buf_competitor).area

intersect_pairs_comp_filtered['comp_area'] = intersect_pairs_comp_filtered.apply(compute_comp_intersection_area, axis=1)
```

```python
comp_overlap_sum = intersect_pairs_comp_filtered.groupby('id_left')['comp_area'].sum().reset_index()
comp_overlap_sum.columns = ['id', 'comp_overlap_area']
```

```python
geometry_target = geometry_target.merge(comp_overlap_sum, on='id', how='left').fillna(0)
```

```python
geometry_target['comp_overlap_ratio'] = geometry_target['comp_overlap_area'] / geometry_target['buffer_area']
```

```python
print("\n目标品牌内部重叠度统计：")
print(geometry_target['overlap_ratio'].describe())
print("\n目标品牌与竞品重叠度统计：")
print(geometry_target['comp_overlap_ratio'].describe())
```
