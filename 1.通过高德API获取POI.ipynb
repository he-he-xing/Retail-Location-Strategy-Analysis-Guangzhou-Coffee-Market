```python
import requests
import json
import pandas as pd
import time
```

```python
api_key='*******************'
url='https://restapi.amap.com/v3/place/polygon'
```

**使用高德地图坐标拾取器获取广州市4个中心城区的经纬度坐标对:**

```python
rectangles = [
    '113.265,23.100|113.442,23.246',  # 天河区 (左下|右上)
    '113.242,23.102|113.301,23.151',  # 越秀区
    '113.233,23.050|113.383,23.267',  # 海珠区
    '113.197,23.027|113.257,23.127'   # 荔湾区
]
```

**要搜索的POI类型有：**

```python
poi_types={
    '咖啡厅':'050500',
    '地铁出入口':'150501',
    '商务写字楼':'120201',
    '购物中心':'060101'
    
}
```

```python
all_pois=[]
for rect in rectangles:
    for type_name,type_code in poi_types.items():
        print(f'''正在搜索矩形{rect}的{type_name}...''')
        page=1
        while True:
            params={
                'key':api_key,
                'polygon':rect,
                'types':type_code,
                'offset':20,
                'page':page,
                'output':'json'
            }
            response=requests.get(url,params=params).json()
            if response.get('status')=='1' :
                pois=response.get('pois',[])
                all_pois.extend(pois)
                if len(pois)<20 :
                    break
                page+=1
            else:
                print(f'''请求失败:{response.get('info')}''')
                break
            time.sleep(0.5)
        time.sleep(0.5)        
```

**将POI列表去重、转为DataFrame,并保存为csv文件：**

```python
df = pd.DataFrame(all_pois).drop_duplicates('id')
df.to_csv('my_selected_areas.csv',index=False,encoding='utf-8')
print(f'''完成，共获取 {len(df)} 条POI''')
```
